```python
import pandas as pd
```

```python
MOI = pd.read_csv('https://raw.githubusercontent.com/AntoineTrabia/Nina_Research_Project/main/MOI.csv')
haploinsufficiency = pd.read_csv('https://raw.githubusercontent.com/AntoineTrabia/Nina_Research_Project/main/haploinsufficiency.csv')
gene_list = pd.read_excel('https://raw.githubusercontent.com/AntoineTrabia/Nina_Research_Project/main/Gene_List.xlsx')
```

```python
list_of_genes = gene_list['Gene Symbol'].unique()
```

```python
MOI = MOI.rename(columns={'gene_symbol': 'Gene Symbol'})
MOI = MOI.loc[:,['Gene Symbol', 'moi_title']]
MOI.drop_duplicates(inplace=True)
MOI_filtered = MOI.loc[MOI['Gene Symbol'].isin(list_of_genes)]
```

```python
haploinsufficiency["Gene Symbol /Region Name"] = haploinsufficiency["Gene Symbol /Region Name"].str.replace(r"HGNC:.*", "", regex=True)
haploinsufficiency = haploinsufficiency.rename(columns={'Gene Symbol /Region Name': 'Gene Symbol'})
haploinsufficiency = haploinsufficiency.loc[:,['Gene Symbol', 'HI Score']]
haploinsufficiency_filtered = haploinsufficiency.loc[haploinsufficiency['Gene Symbol'].isin(list_of_genes)]
```

```python
merged_df = pd.merge(MOI, haploinsufficiency, on=['Gene Symbol'], how='outer')
merged_df
```

```python
merged_df_filtered = pd.merge(MOI_filtered, haploinsufficiency_filtered, on=['Gene Symbol'], how='outer')
merged_df_filtered
```

```python
haploinsufficiency_filtered.to_excel('haploinsufficiency.xlsx',index=False)
MOI_filtered.to_excel('MOI.xlsx',index=False)
merged_df.to_excel('Total.xlsx',index=False)
merged_df_filtered.to_excel('filtered.xlsx',index=False)
```
